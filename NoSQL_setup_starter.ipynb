2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

3. Update the new restaurant with the `BusinessTypeID` you found.

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude

# Define the query to find documents with the fields to update
query = {}  # You can add conditions to this query if needed

# Define the new data type for longitude and latitude fields
update_query = {
    "$set": {
        "longitude": Decimal128(str(decimal.Decimal("0.08384000"))),
        "latitude": Decimal128(str(decimal.Decimal("51.49014200")))
    }
}

# Use the $set operator to update the fields with the new data type
result = establishments.update_many(query, update_query)

result.modified_count
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
# Define the update query to set non-rating values to None
update_query_none = [
    {
        "$set": {
            "RatingValue": {
                "$cond": {
                    "if": {"$in": ["$RatingValue", non_ratings]},
                    "then": None,
                    "else": "$RatingValue"
                }
            }
        }
    }
]

# Use the update_many() method to apply the update query to set non-rating values to None
result_none = establishments.update_many({}, update_query_none)

# Define the update query to convert RatingValue to integers
update_query_int = [
    {
        "$set": {
            "RatingValue": {
                "$toInt": "$RatingValue"
            }
        }
    }
]

# Use the update_many() method to apply the update query to convert RatingValue to integers
result_int = establishments.update_many({}, update_query_int)

# Print the number of documents updated for both operations
print(f"Number of documents with non-rating values set to None: {result_none.modified_count}")
print(f"Number of documents with RatingValue converted to integers: {result_int.modified_count}")
```
